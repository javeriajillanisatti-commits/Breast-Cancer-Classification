## Part A – Dataset Preparation

Import Libraries & Load Dataset

```python
import pandas as pd

# Load Dataset
df = pd.read_csv("data.csv")

# Display first 5 rows
df.head()
```

Dataset Information

```python
# Dataset size
print("Dataset Shape:", df.shape)

# Dataset information
df.info()
```

Describe Dataset

```python
# Statistical description
df.describe()
```

Display Features and Target Variable

```python
# Display all columns
print(df.columns.tolist())
```

```python
# Features and Target Variable

X = df.drop("diagnosis", axis=1)
y = df["diagnosis"]

# Display Target Variable
print("Target Variable:")
print(y.name)

# Display Target Classes
print("\nTarget Classes:")
print(y.unique())

# Display Class Distribution
print("\nTarget Value Counts:")
print(y.value_counts())
```

## Preprocess the Dataset

Check Missing Values

```python
# Check missing values
print(df.isnull().sum())
```

Encode Target Variable

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

y = le.fit_transform(y)

print(y[:10])
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X = scaler.fit_transform(X)
```

Split Dataset into Training and Testing Sets

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("Training Features:", X_train.shape)
print("Testing Features:", X_test.shape)
print("Training Labels:", y_train.shape)
print("Testing Labels:", y_test.shape)
```

## Part B – Cross-Validation

```python
from sklearn.impute import SimpleImputer

# Replace missing values with mean
imputer = SimpleImputer(strategy="mean")

X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)

print("Missing values handled successfully")
```

Train SVM Model

```python
from sklearn.svm import SVC

# Create Model
model = SVC(kernel="rbf")

# Train Model
model.fit(X_train, y_train)

print("Model trained successfully")
```

Apply K-Fold Cross-Validation

```python
from sklearn.impute import SimpleImputer
from sklearn.model_selection import KFold, cross_val_score

# Handle missing values in complete dataset
imputer = SimpleImputer(strategy="mean")

X = imputer.fit_transform(X)

# Define K-Fold
kfold = KFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

# Apply Cross Validation
scores = cross_val_score(
    model,
    X,
    y,
    cv=kfold,
    scoring="accuracy"
)

print("Cross Validation Scores:")
print(scores)

print("Mean Accuracy:", scores.mean())
print("Standard Deviation:", scores.std())
```

Interpret Cross-Validation Results

```python
# Calculate Cross Validation Results

mean_accuracy = scores.mean()
std_accuracy = scores.std()

print("Mean Accuracy:", mean_accuracy)
print("Standard Deviation:", std_accuracy)
```

## Part C – Hyperparameter Tuning

Optimize Model using GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC

# Define SVM Model
svm = SVC()

# Define Parameters
param_grid = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}

# Apply GridSearchCV
grid_search = GridSearchCV(
    svm,
    param_grid,
    cv=5,
    scoring="accuracy"
)

# Train Grid Search
grid_search.fit(X_train, y_train)

print("Best Parameters:")
print(grid_search.best_params_)

print("Best Accuracy:")
print(grid_search.best_score_)
```

Optimize Same Model using RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV

# Define Parameters
param_random = {
    'C': [0.01, 0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto']
}

# Apply RandomizedSearchCV
random_search = RandomizedSearchCV(
    svm,
    param_random,
    n_iter=10,
    cv=5,
    scoring="accuracy",
    random_state=42
)

# Train Random Search
random_search.fit(X_train, y_train)

print("Best Parameters:")
print(random_search.best_params_)

print("Best Accuracy:")
print(random_search.best_score_)
```

Compare Optimal Parameters

```python
print("GridSearch Best Parameters:")
print(grid_search.best_params_)

print("\nRandomizedSearch Best Parameters:")
print(random_search.best_params_)
```

Evaluate Best Model

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix
)

# Select Best Model
best_model = grid_search.best_estimator_

# Prediction
y_pred = best_model.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))

print("Precision:", precision_score(y_test, y_pred))

print("Recall:", recall_score(y_test, y_pred))

print("F1 Score:", f1_score(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Generate Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

# Plot Figure
plt.figure(figsize=(6,4))

sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=["Benign", "Malignant"],
    yticklabels=["Benign", "Malignant"]
)

plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")
plt.title("Confusion Matrix - SVM Model")

plt.show()
```
